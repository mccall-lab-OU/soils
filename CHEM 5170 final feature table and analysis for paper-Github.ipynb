```r
#random forest analysis - necessary package
install.packages("randomForest", repos = "http://cran.us.r-project.org")
```

```r
#random forest analysis to identify metabolites that differ between Binger, Norman and Oklahoma City, OK
#correcting for different sampling sizes between cities
OK_cities_moreThan5<-droplevels(subset(features, features$ATTRIBUTE_City_Collection==c("Binger")|features$ATTRIBUTE_City_Collection==c("Norman")|features$ATTRIBUTE_City_Collection==c("Oklahoma City")))
library("randomForest")
rffit1 <- randomForest(OK_cities_moreThan5$ATTRIBUTE_City_Collection ~., data= OK_cities_moreThan5[,40:3841], importance=TRUE, ntree=10000, na.omit=TRUE, strata=OK_cities_moreThan5$ATTRIBUTE_City_Collection, sampsize=rep(min(table(OK_cities_moreThan5$ATTRIBUTE_City_Collection)), nlevels(OK_cities_moreThan5$ATTRIBUTE_City_Collection)))
rffit1
varImp<-varImpPlot(rffit1, type=1)
write.csv(rffit1$importance, "varImpRF_OKCitiesMoreThan5_Github.csv")
```
